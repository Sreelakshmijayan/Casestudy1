# 3. Plot a bar graph of male vs female buyers participated in the sale

# 4.Find the top 5 cars based on their sales price

```python
data1.Make.value_counts().nlargest(5).plot(kind='bar',figsize=(10,5))
plt.title('Growth Index')
plt.xlabel('Make')
plt.ylabel('Sale Price')
```

# 5. Find least 5 cars based on their Resell price

```python
data1.Make.value_counts().nsmallest(5).plot(kind='bar',figsize=(10,5))
plt.title('Growth Index')
plt.xlabel('Make')
plt.ylabel('Resell Price')
```
